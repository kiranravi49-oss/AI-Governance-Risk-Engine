```python
from sklearn.metrics import recall_score

male_mask = gender_test == "Male"
female_mask = gender_test == "Female"

print("Male samples:", male_mask.sum())
print("Female samples:", female_mask.sum())

print("Male Recall:", recall_score(y_test[male_mask], y_pred[male_mask]))
print("Female Recall:", recall_score(y_test[female_mask], y_pred[female_mask]))
```

```python
print(gender_test.unique())
print(gender_test.value_counts())
```

```python
import numpy as np

male_positive_rate = np.mean(y_pred[male_mask])
female_positive_rate = np.mean(y_pred[female_mask])

print("Male Positive Prediction Rate:", male_positive_rate)
print("Female Positive Prediction Rate:", female_positive_rate)
```

```python
fairness_gap = abs(male_positive_rate - female_positive_rate)

print("Fairness Gap:", fairness_gap)
```

```python
if fairness_gap < 0.05:
    fairness_score = 100
elif fairness_gap < 0.10:
    fairness_score = 75
elif fairness_gap < 0.20:
    fairness_score = 50
else:
    fairness_score = 25

print("Fairness Score (out of 100):", fairness_score)
```

```python
accuracy_score_value = accuracy_score(y_test, y_pred)

if accuracy_score_value >= 0.90:
    accuracy_score_gov = 100
elif accuracy_score_value >= 0.85:
    accuracy_score_gov = 80
elif accuracy_score_value >= 0.80:
    accuracy_score_gov = 60
else:
    accuracy_score_gov = 40

print("Governance Accuracy Score:", accuracy_score_gov)
```

```python
imbalance_ratio = y.value_counts(normalize=True).min()

if imbalance_ratio >= 0.40:
    imbalance_score = 100
elif imbalance_ratio >= 0.30:
    imbalance_score = 80
elif imbalance_ratio >= 0.20:
    imbalance_score = 60
else:
    imbalance_score = 40

print("Imbalance Score:", imbalance_score)
```

```python
governance_index = (
    0.4 * accuracy_score_gov +
    0.4 * fairness_score +
    0.2 * imbalance_score
)

print("Final Governance Index:", governance_index)
```

```python
if governance_index >= 85:
    risk_level = "Low Risk – Production Ready"
elif governance_index >= 70:
    risk_level = "Moderate Risk – Monitor Closely"
elif governance_index >= 50:
    risk_level = "High Risk – Mitigation Required"
else:
    risk_level = "Critical Risk – Do Not Deploy"

print("Governance Risk Level:", risk_level)
```
